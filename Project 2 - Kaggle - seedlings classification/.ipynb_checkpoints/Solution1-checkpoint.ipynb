## Feature extraction - Naive approach

### SIFT

```python
# extracting SIFT keypoints and descriptors for one image
def SIFTFeature(fileImage):
    #open image
    img = cv2.imread(fileImage)
    
    #convert to gray
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #create SIFT object
    sift = cv2.xfeatures2d.SIFT_create()
    keypoints, descriptors = sift.detectAndCompute(gray,None)
    
    return keypoints, descriptors
```

```python
# function for displaying image
def imshow(img, hight=640, width=480, name="me",wait=0):
    cv2.namedWindow(name, cv2.WINDOW_NORMAL)
    cv2.resizeWindow(name, hight, width)
    cv2.imshow(name, img)
    cv2.waitKey(wait)
    cv2.destroyAllWindows()
```

Problem are lots of pixels that do not belong to plant. Method such as largest connected component can be used to remove artifacts.

### Extracting largest connected component

```python
# connected components
nlabels, labelsMap, stats, centroids = cv2.connectedComponentsWithStats(th)
""" 
stats, for every component:
(1) x coordinate of bounding box, 
(2) y coordinate of bounding box
(3) horizontal length of bounding box,
(3) verical length of bounding box,
(4) total number of pixels of component (area)
"""

print("no components:" + str(nlabels))
print()

labels, count = np.unique(labelsMap, return_counts=True)
print("component labels:") 
print(labels)
print()
print("Number of pixels per component:")
print(count)
print()
print(stats[:,4])
```

```python
# area of all components (number of pixels)
area = stats[:,4]
# first element is number of pixels for all components
area[0] = 0 
# find label with most pixels (this will be plant)
max_label = np.argmax(area)
print(max_label)

# extract largest component
img2 = np.zeros(labelsMap.shape)
img2[labelsMap == max_label] = 255

# important: for openCV functions use uint8!
img2 = img2.astype(np.uint8)
print(type(img2[0][0]))
#display
print(img2)

plt.imshow(img2)
plt.show()
```

### HOG

```python
# HOG descriptor
def hog(img):

    grad_x = cv2.Sobel(img, cv2.CV_32F, 1, 0)  
    grad_y = cv2.Sobel(img, cv2.CV_32F, 0, 1)

    bin_n = 16
    magnitude, angle = cv2.cartToPolar(grad_x, grad_y)
    bins = np.int32(bin_n*angle/(2*np.pi))
    bin_cells = bins[:10,:10], bins[10:,:10], bins[:10,10:], bins[10:,10:]
    mag_cells = magnitude[:10,:10], magnitude[10:,:10], magnitude[:10,10:], magnitude[10:,10:]

    hists = [np.bincount(b.ravel(), m.ravel(), bin_n) for b, m in zip(bin_cells, mag_cells)]
    hist = np.hstack(hists)     # hist is a 64 bit vector
    
    return hist

hog(img2)
```

## Feature preparation
* for all images:
 * RGB to HSV and use only H
 * Segment the H (otsu, H is bimodal in this case)
 * Remove artifacts (largest connected component)
 * use descriptor for regions (segmented image): moments

```python
def cvtRGB2HSVExtH(img):
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return img_hsv[:,:,0]

def blurSegmentBimodal(img):
    # Otsu's thresholding after Gaussian filtering
    blur = cv2.GaussianBlur(img,(5,5),0)
    thresh_val, thresholded = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    return thresholded

def connectedComponent(img):
    nlabels, labelsMap, stats, centroids = cv2.connectedComponentsWithStats(img)
    
    area = stats[:,4]
    area[0] = 0 
    max_label = np.argmax(area)

    img2 = np.zeros(labelsMap.shape)
    img2[labelsMap == max_label] = 255

    img2 = img2.astype(np.uint8)
    
    return img2

def momentFeatures(img):
    return cv2.HuMoments(cv2.moments(img)).flatten()

def img2Feat_pipeline(img):
    H = cvtRGB2HSVExtH(img)
    thresh = blurSegmentBimodal(H)
    largest_comp = connectedComponent(thresh)
    feat = momentFeatures(largest_comp)
    return feat
```

```python
def y_path_dict2y_feat(y_path_dict):
    y_feat = {}
    for key, vals in y_path_dict.items():
        feat_val = np.empty(shape=(len(vals),7)) # hu returns 7 moments per image
        i = 0
        for val in vals:
            path = trainFolder + '/' + yMap[key] + '/' + val
            img = cv2.imread(path)
            feat_val[i] =  img2Feat_pipeline(img)
            i = i + 1
        y_feat[key] = feat_val
    return y_feat
```

```python
train = y_path_dict2y_feat(trainSet)
```

```python
def test2feat(test):
    X_test = np.empty(shape=(len(test),7))
    i = 0
    for filename in test:
        path = testFolder + '/' + filename
        img = cv2.imread(path)
        X_test[i] = img2Feat_pipeline(img)
        i = i + 1
    return X_test
```
